```python
# Трейдер хочет, чтобы программа проверяла, получил ли он прибыль или убы­ток по сделке. Напишите программу, в которой с клавиатуры вводится цена по­купки и продажи, а в ответ программа сообщает, получил ли трейдер прибыль.
purchase = float(input("Purchase sum is: "))
sale = float(input("Sale sum is: "))
profit_or_loss = sale - purchase
if profit_or_loss > 0:
    print(f"The trader gets profit: {profit_or_loss:2f}")
elif profit_or_loss < 0:
    print(f"The trader gets loss: {abs(profit_or_loss):.2f}")
else:
    print("Profit is zero")
```

```python
# Трое сотрудников, Самир, Правин и Мохит, вводят свой опыт работы с клавиа­ туры. Напишите программу, которая определяет наиболее и наименее опытного из них.
emploies = dict()

emploies["Самир"] = int(input("Опыт работы Самира: "))
emploies["Правин"] = int(input("Опыт работы Правина: "))
emploies["Мохит"] = int(input("Опыт работы Мохита: "))

print(f"Самый опытный - {max(emploies.items(), key=lambda t: t[1])[0]}")
print(f"Самый неопытный - {min(emploies.items(), key=lambda t: t[1])[0]}")
```

```python
# Сумма трех углов треугольника составляет 1 80 . Напишите программу, которая проверяет, является ли фигура треугольником, если в качестве входных данных вводятся три угла.
first = int(input("Введите величину первого угла: "))
second = int(input("Введите величину второго угла: "))
third = int(input("Введите виличину третьего угла: "))
if first + second + third == 180:
    print("Фигура является  треугольником")
else:
    print("Фигура не является треугольником")
```

```python
# Даны три угла треугольника. Напишите программу, которая проверит, является ли он прямоугольным (один из углов должен составлять 90°).
first = int(input("Введите величину первого угла: "))
second = int(input("Введите величину второго угла: "))
third = int(input("Введите величину третьего угла: "))
if (
    first > 0
    and second > 0
    and third > 0
    and sum([first, second, third]) == 180
):
    if first == 90 or second == 90 or third == 90:
        print("Фигура  является четырехугольным треугольником")
    else:
        print("Фигура не является четырехугольным треугольником")
else:
    print("Фигура не является треугольником")
```

```python
# Напишите программу для вывода всех простых чисел от 1 до 500.
def simple_nums(a, b):
    for i in range(a, b + 1):
        if i == 1:
            continue

        deviders = 0
        for j in range(1, i + 1):
            if i % j == 0:
                deviders += 1

        if deviders == 2:
            yield i


print(", ".join(str(x) for x in simple_nums(1, 500)))
```

```python
# Выведите все числа, кратные 9, которые меньше 300 (подсказка: вы можете ис­пользовать функцию range ( )  .
def nums():
    for i in range(301):
        if i % 9 == 0:
            yield i


print(",".join(str(x) for x in nums()))
```
