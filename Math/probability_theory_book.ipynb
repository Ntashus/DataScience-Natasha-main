### 1.2. Операции над множествами  

1. Дополнение.  Ac  - множество точек, которые не входят в A, исключительно точки пространства Ω  
Ac ={ω|ω∈/A}  «Ac — это множество точек ω, не принадлежащих множеству A» Ωc = ∅ и ∅c = Ω   

Операция взятия дополнения обладает следующим свойством: если ее применить дважды, то мы снова получим множество A:
(Ac)c = A  

2. Объединение  
Объединением A∪B множеств A и B называется мно- жество точек, которые принадлежат хотя бы одному из этих множеств. Символически это записывается так:
 A∪B={ω|ω∈A или ω∈B}  

 3. Пересечение  

 Пересечением A∩B множеств A и By называется мно- жество точек, принадлежащих обоим множествам. В символической форме:
A∩B={ω|ω∈A и ω∈B}.  

**Коммутативный закон**: A∪B=B∪A, A∩B=B∩A.
**Ассоциативный закон**: (A∪B)∪C =A∪(B∪C); (A ∩ B) ∩ C = A ∩ (B ∩ C)  

**Дистрибутивные законы**  

(A∪B)∩C =(A∩C)∪(B∩C); 

(A∩B)∪C =(A∪C)∩(B∪C).  

**Законы Де Моргана**  

(A ∪ B)c = Ac ∩ Bc;  
 (A ∩ B)c = Ac ∪ Bc  

 A ∩ B = (Ac ∪ Bc)c  
 A ∪ B = (Ac ∩ Bc)c  

 4. Разность   

 Множество A\B содержит точки, принадлежащие A и (но) не принадлежащие B. В символической форме:
A\B=A∩Bc={ω|ω∈A и ω∈/B}  

5. Симметрическая разность  
Множество A△B содержит точки, при- надлежащие в точности одному из двух множеств A и B. В символиче- ской форме это можно выразить так:
A△B=(A∩Bc)∪(Ac ∩B)=(A\B)∪(B\A).
